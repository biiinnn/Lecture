## Preprocessing toy example 과제 _ 17101971 이예빈

```python
import nltk
from pprint import pprint
```

Text Data: 2005년 스탠포드 대학 스티브잡스 연설문

```python
doc = open('SteveJobs.txt', 'r', encoding='utf-8').read()
doc = doc.replace('\n', ' ')
doc
```

### 1. Tokenizing

```python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize, wordpunct_tokenize

sent_tokenize_result = sent_tokenize(doc)
word_tokenize_result = word_tokenize(doc)
wordpunct_tokenize_result = wordpunct_tokenize(doc)

print("==== Result of sent_tokenize ====")
pprint(sent_tokenize_result)
print("\n")
print("==== Result of word_tokenize ====")
pprint(word_tokenize_result)
print("\n")
print("==== Result of wordpunct_tokenize ====")
pprint(wordpunct_tokenize_result)
```

### 2. POS tagging

```python
pos_result = nltk.pos_tag(wordpunct_tokenize_result)
pprint(pos_result)
```

### 3. Stopword 제거

```python
import string
from nltk.corpus import wordnet as wn
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer

lemmatizer  = WordNetLemmatizer()
stopwords   = set(stopwords.words('english'))
punctuation = string.punctuation
```

```python
punctuation
```

```python
print(stopwords)
```

```python
def remove_stopwords(text):
    for token, tag in pos_result:
        token = token.lower()
        if token in stopwords or token in punctuation:
            continue
        token = lemmatizer.lemmatize(token)
        yield token
        
stopwords_removed = list(remove_stopwords(doc))

print(stopwords_removed)
```

### 4. Stemming

```python
from nltk.stem.snowball import SnowballStemmer
from nltk.stem.lancaster import LancasterStemmer
from nltk.stem.porter import PorterStemmer

snowball = SnowballStemmer('english')
lancaster = LancasterStemmer()
porter = PorterStemmer()

for stemmer in (snowball, lancaster, porter):
    stemmed_text = [stemmer.stem(t) for t in stopwords_removed]
    print(" ".join(stemmed_text))
    print(stemmed_text)
```

### 5. 빈도수 시각화

```python
tokens = nltk.Text(stemmed_text)
```

```python
print(len(tokens))       # 토큰의 개수
print(len(set(tokens)))  # 중복제거
tokens.vocab()                  # 빈도수
```

```python
tokens.vocab().most_common(30)  #상위 30개
```
