## http://media.daum.net/digital 에서 뉴스 리스트와 URL을 불러오기

```python
import requests
from bs4 import BeautifulSoup
```

```python
res = requests.get('http://media.daum.net/digital')
```

```python
print(res.content)
```

```python
soup = BeautifulSoup(res.content, 'html.parser')
```

```python
soup
```

```python
soup.body
```

```python
type(soup.body)
```

### 주요기사 제목

1. main1: 가장 큰 기사 제목
2. main2: main1 밑 4개 기사 제목
3. main3: main1,main2 옆 5개 기사 제목

```python
main1_div_1 = soup.body.find_all('div', class_='cont_category')[0]
main1_div_1
```

```python
main1_div_2 = main1_div_1.find_all('div', id="cSub")[0]
main1_div_2
```

```python
main1_div_3 = main1_div_2.find_all('div', class_="cate_sub")[0]
main1_div_3
```

```python
main1_div_4 = main1_div_3.find_all('div', class_="section_cate section_headline")[0]
main1_div_4
```

```python
main1_div_5 = main1_div_4.find_all('div', class_="item_mainnews")[0]
main1_div_5
```

```python
main1_div_6 = main1_div_5.find_all('div', class_="cont_thumb")[0]
main1_div_6
```

```python
main1_list = main1_div_6.find_all('strong', class_="tit_thumb")
main1_list
```

```python
main_news_list=[]
for item in main1_list:
    main_news_list.append(item.text.strip())
```

```python
main_news_list
```

```python
main_final_url = []
for item in main1_list:
    main_final_url.append(item.a["href"])
```

```python
main_final_url
```

```python
#이전까지 경로는 main1의 경로와 동일

main2_div_1 = main1_div_4.find_all('ul', class_="list_mainnews")[0]
main2_div_1
```

```python
main2_list= main2_div_1.find_all('li')
main2_list
```

```python
for item in main2_list:
    main_news_list.append(item.div.strong.text.strip())
```

```python
main_news_list
```

```python
for item in main2_list:
    main_final_url.append(item.a["href"])
```

```python
main_final_url
```

```python
#이전까지 경로는 main1의 경로와 동일

main3_div_1 = main1_div_4.find_all('ul', class_="list_mainnews2")[0]
main3_div_1
```

```python
main3_list= main3_div_1.find_all('li')
main3_list
```

```python
for item in main3_list:
    main_news_list.append(item.a.text.strip())
```

```python
main_news_list
```

```python
for item in main3_list:
    main_final_url.append(item.a["href"])
```
